```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import transforms, datasets, models
from torch.utils.data import DataLoader, Subset
import os
from tqdm import tqdm
import warnings
from collections import Counter
warnings.filterwarnings('ignore')

# Set random seeds
torch.manual_seed(42)
np.random.seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# ==================== OPTIMIZED DATA PREPARATION ====================
print("=" * 60)
print("FAST TRAINING - OPTIMIZED FOR 25 EPOCHS")
print("=" * 60)

# Balanced augmentation - not too weak, not too strong
train_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),  # Moderate cropping
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(15),  # Moderate rotation
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),  # Lighter color jitter
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])

val_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load dataset
full_dataset = datasets.ImageFolder(
    root="/kaggle/input/real-and-fake-face-detection/real_and_fake_face"
)

# Stratified split
all_labels = [label for _, label in full_dataset.samples]
train_indices, val_indices = train_test_split(
    list(range(len(full_dataset))),
    test_size=0.15,
    stratify=all_labels,
    random_state=42
)

train_dataset = Subset(full_dataset, train_indices)
val_dataset = Subset(full_dataset, val_indices)

# Apply transforms
train_dataset.dataset.transform = train_transform
val_dataset.dataset.transform = val_transform

# Larger batch size for faster training
batch_size = 64  # Increased from 32
num_workers = 4

train_loader = DataLoader(
    train_dataset, 
    batch_size=batch_size, 
    shuffle=True,
    num_workers=num_workers, 
    pin_memory=True,
    drop_last=True
)

val_loader = DataLoader(
    val_dataset, 
    batch_size=batch_size, 
    shuffle=False,
    num_workers=num_workers, 
    pin_memory=True
)

print(f"Training samples: {len(train_dataset)}")
print(f"Validation samples: {len(val_dataset)}")
print(f"Classes: {full_dataset.classes}")
print(f"Class balance: {Counter([full_dataset.targets[i] for i in train_indices])}")
print(f"Batch size: {batch_size} (larger = faster training)")
print(f"Total batches per epoch: {len(train_loader)}")

# ==================== HIGH-PERFORMANCE MODEL ====================
print("\n" + "=" * 60)
print("BUILDING HIGH-PERFORMANCE MODEL")
print("=" * 60)

class FastAccurateFaceDetector(nn.Module):
    """
    Optimized for speed AND accuracy:
    - ResNet50 pretrained backbone
    - Efficient classifier head
    - Proper dropout for regularization
    """
    def __init__(self, num_classes=2, pretrained=True):
        super(FastAccurateFaceDetector, self).__init__()
        
        # ResNet50 - excellent balance of speed and accuracy
        self.backbone = models.resnet50(pretrained=pretrained)
        num_features = self.backbone.fc.in_features  # 2048
        
        # Remove original FC
        self.backbone.fc = nn.Identity()
        
        # Optimized classifier - deeper but efficient
        self.classifier = nn.Sequential(
            nn.Dropout(p=0.5),
            nn.Linear(num_features, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(inplace=True),
            
            nn.Dropout(p=0.4),
            nn.Linear(1024, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(inplace=True),
            
            nn.Dropout(p=0.3),
            nn.Linear(512, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(inplace=True),
            
            nn.Linear(128, num_classes)
        )
        
        self._initialize_weights()
        
    def _initialize_weights(self):
        """Proper weight initialization for faster convergence"""
        for m in self.classifier.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm1d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)
    
    def forward(self, x):
        features = self.backbone(x)
        output = self.classifier(features)
        return output

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
if device.type == 'cpu':
    print("⚠️  WARNING: Training on CPU will be VERY slow!")
    print("   Enable GPU in Kaggle: Settings → Accelerator → GPU T4 x2")

model = FastAccurateFaceDetector(num_classes=2, pretrained=True)
model = model.to(device)

print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
print(f"Trainable parameters: {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")

# ==================== AGGRESSIVE TRAINING STRATEGY ====================
print("\n" + "=" * 60)
print("SETUP: AGGRESSIVE TRAINING FOR FAST CONVERGENCE")
print("=" * 60)

# Loss with label smoothing
class LabelSmoothingLoss(nn.Module):
    def __init__(self, classes=2, smoothing=0.1):
        super().__init__()
        self.confidence = 1.0 - smoothing
        self.smoothing = smoothing
        self.classes = classes
        
    def forward(self, pred, target):
        pred = pred.log_softmax(dim=-1)
        with torch.no_grad():
            true_dist = torch.zeros_like(pred)
            true_dist.fill_(self.smoothing / (self.classes - 1))
            true_dist.scatter_(1, target.data.unsqueeze(1), self.confidence)
        return torch.mean(torch.sum(-true_dist * pred, dim=-1))

criterion = LabelSmoothingLoss(classes=2, smoothing=0.1)

# Freeze backbone initially for faster Phase 1 training
def freeze_backbone(model):
    for param in model.backbone.parameters():
        param.requires_grad = False
    for param in model.classifier.parameters():
        param.requires_grad = True

def unfreeze_all(model):
    for param in model.parameters():
        param.requires_grad = True

freeze_backbone(model)
print("Phase 1: Training classifier only (faster convergence)")

# AGGRESSIVE optimizer for Phase 1
optimizer = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, model.parameters()),
    lr=3e-3,  # Higher learning rate for faster learning
    weight_decay=1e-4
)

# One-Cycle Learning Rate for maximum efficiency
from torch.optim.lr_scheduler import OneCycleLR

total_epochs = 25  # TARGET: 25 EPOCHS ONLY!
steps_per_epoch = len(train_loader)

scheduler = OneCycleLR(
    optimizer,
    max_lr=3e-3,
    epochs=6,  # Phase 1: 6 epochs
    steps_per_epoch=steps_per_epoch,
    pct_start=0.3,
    anneal_strategy='cos'
)

print(f"Total target epochs: {total_epochs}")
print(f"Phase 1 (frozen): 6 epochs with high LR (3e-3)")
print(f"Phase 2 (unfrozen): 19 epochs with lower LR (1e-4)")

# ==================== TRAINING FUNCTIONS ====================

def train_epoch(model, loader, optimizer, criterion, scheduler, device, use_scheduler=True):
    """Efficient training loop"""
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    pbar = tqdm(loader, desc='Training', leave=False)
    for images, labels in pbar:
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        
        # Gradient clipping for stability
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        optimizer.step()
        if use_scheduler:
            scheduler.step()
        
        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)
        
        pbar.set_postfix({
            'loss': f'{loss.item():.4f}',
            'acc': f'{100*correct/total:.2f}%'
        })
    
    return running_loss / len(loader), 100 * correct / total

def validate(model, loader, criterion, device):
    """Fast validation"""
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for images, labels in tqdm(loader, desc='Validating', leave=False):
            images, labels = images.to(device), labels.to(device)
            
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            
            correct += (predicted == labels).sum().item()
            total += labels.size(0)
            
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    return (running_loss / len(loader), 
            100 * correct / total, 
            all_preds, all_labels)

# ==================== MAIN TRAINING LOOP ====================
print("\n" + "=" * 60)
print("STARTING FAST TRAINING - TARGET: 95% IN 25 EPOCHS")
print("=" * 60)

history = {
    'train_loss': [], 'train_acc': [],
    'val_loss': [], 'val_acc': [],
    'lr': []
}

best_val_acc = 0.0
patience_counter = 0
early_stop_patience = 10
phase_1_epochs = 6

for epoch in range(1, total_epochs + 1):
    
    # PHASE 2: Unfreeze and fine-tune
    if epoch == phase_1_epochs + 1:
        print("\n" + "=" * 60)
        print(f"PHASE 2: UNFREEZING BACKBONE (Epoch {epoch})")
        print("=" * 60)
        unfreeze_all(model)
        
        # New optimizer with differential learning rates
        optimizer = torch.optim.AdamW([
            {'params': model.backbone.parameters(), 'lr': 1e-4},  # Low LR for backbone
            {'params': model.classifier.parameters(), 'lr': 5e-4}  # Higher LR for classifier
        ], weight_decay=1e-4)
        
        # Cosine annealing for Phase 2
        scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
            optimizer,
            T_max=total_epochs - phase_1_epochs,
            eta_min=1e-6
        )
        
        print("✓ Using differential learning rates")
        print("  Backbone:   1e-4")
        print("  Classifier: 5e-4")
    
    # Train
    use_scheduler = epoch <= phase_1_epochs
    train_loss, train_acc = train_epoch(
        model, train_loader, optimizer, criterion, 
        scheduler if use_scheduler else None, 
        device, use_scheduler
    )
    
    # Validate
    val_loss, val_acc, val_preds, val_labels = validate(
        model, val_loader, criterion, device
    )
    
    # Update scheduler (Phase 2 only)
    if epoch > phase_1_epochs:
        scheduler.step()
    
    current_lr = optimizer.param_groups[0]['lr']
    
    # History
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)
    history['lr'].append(current_lr)
    
    # Print progress
    phase = "PHASE 1" if epoch <= phase_1_epochs else "PHASE 2"
    print(f"\n[{phase}] Epoch {epoch}/{total_epochs}:")
    print(f"  Train: Loss={train_loss:.4f}, Acc={train_acc:.2f}%")
    print(f"  Val:   Loss={val_loss:.4f}, Acc={val_acc:.2f}%")
    print(f"  LR: {current_lr:.8f}")
    
    # Save best model
    if val_acc > best_val_acc:
        improvement = val_acc - best_val_acc
        best_val_acc = val_acc
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_acc': val_acc,
        }, 'best_fast_model.pth')
        print(f"  ✓ BEST! Val Acc: {val_acc:.2f}% (+{improvement:.2f}%)")
        patience_counter = 0
        
        # Check if target reached
        if val_acc >= 95.0:
            print(f"\n{'='*60}")
            print(f"🎉 TARGET REACHED! 95%+ accuracy at epoch {epoch}!")
            print(f"{'='*60}")
    else:
        patience_counter += 1
        print(f"  No improvement ({patience_counter}/{early_stop_patience})")
    
    # Early stopping
    if patience_counter >= early_stop_patience:
        print(f"\nEarly stopping at epoch {epoch}")
        break
    
    # Progress milestones
    if epoch == phase_1_epochs:
        print(f"\n{'='*60}")
        print(f"PHASE 1 COMPLETE - Validation Acc: {val_acc:.2f}%")
        print(f"Expected: 75-85% | Actual: {val_acc:.2f}%")
        if val_acc < 70:
            print("⚠️  Lower than expected - may need more epochs")
        print(f"{'='*60}")
    
    if epoch == 15:
        print(f"\n{'='*60}")
        print(f"CHECKPOINT - Epoch 15 | Val Acc: {val_acc:.2f}%")
        print(f"Expected: 88-93% | Actual: {val_acc:.2f}%")
        if val_acc >= 95:
            print("🎉 Target reached early!")
        elif val_acc >= 90:
            print("✓ On track for 95%+")
        else:
            print("⚠️  May not reach 95% in 25 epochs")
        print(f"{'='*60}")
    
    print("-" * 60)

print("\n" + "=" * 60)
print(f"TRAINING COMPLETED!")
print(f"Best Validation Accuracy: {best_val_acc:.2f}%")
print(f"Total Epochs: {len(history['train_loss'])}")
print("=" * 60)

# ==================== FINAL EVALUATION ====================
print("\nLoading best model for final evaluation...")
checkpoint = torch.load('best_fast_model.pth')
model.load_state_dict(checkpoint['model_state_dict'])

val_loss, val_acc, val_preds, val_labels = validate(
    model, val_loader, criterion, device
)

print(f"\n{'='*60}")
print(f"FINAL RESULTS")
print(f"{'='*60}")
print(f"Validation Accuracy: {val_acc:.2f}%")
print(f"Validation Loss: {val_loss:.4f}")

if val_acc >= 95:
    print(f"\n🎉 SUCCESS! Achieved {val_acc:.2f}% accuracy!")
elif val_acc >= 90:
    print(f"\n✓ Good! Achieved {val_acc:.2f}% accuracy")
    print("To reach 95%: train 5-10 more epochs or use ensemble")
else:
    print(f"\n⚠️  Achieved {val_acc:.2f}% accuracy")
    print("Recommendations:")
    print("  1. Enable GPU for faster/better training")
    print("  2. Train for 30-35 epochs instead of 25")
    print("  3. Check data quality")

print("\nDetailed Classification Report:")
print(classification_report(
    val_labels, val_preds,
    target_names=full_dataset.classes,
    digits=4
))

# ==================== VISUALIZATIONS ====================
os.makedirs('results', exist_ok=True)

# Compact training history
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Loss
axes[0].plot(history['train_loss'], 'o-', label='Train', linewidth=2)
axes[0].plot(history['val_loss'], 's-', label='Val', linewidth=2)
axes[0].axvline(x=phase_1_epochs-1, color='r', linestyle='--', alpha=0.5, label='Unfreeze')
axes[0].set_xlabel('Epoch', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Loss', fontsize=12, fontweight='bold')
axes[0].set_title('Training Loss', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Accuracy
axes[1].plot(history['train_acc'], 'o-', label='Train', linewidth=2)
axes[1].plot(history['val_acc'], 's-', label='Val', linewidth=2)
axes[1].axhline(y=95, color='g', linestyle='--', alpha=0.7, linewidth=2, label='Target 95%')
axes[1].axvline(x=phase_1_epochs-1, color='r', linestyle='--', alpha=0.5, label='Unfreeze')
axes[1].set_xlabel('Epoch', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Accuracy (%)', fontsize=12, fontweight='bold')
axes[1].set_title('Training Accuracy', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].set_ylim([0, 105])

# Learning rate
axes[2].plot(history['lr'], 'o-', linewidth=2, color='green')
axes[2].axvline(x=phase_1_epochs-1, color='r', linestyle='--', alpha=0.5, label='Unfreeze')
axes[2].set_xlabel('Epoch', fontsize=12, fontweight='bold')
axes[2].set_ylabel('Learning Rate', fontsize=12, fontweight='bold')
axes[2].set_title('Learning Rate Schedule', fontsize=14, fontweight='bold')
axes[2].set_yscale('log')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('results/fast_training_history.png', dpi=300, bbox_inches='tight')
print("\n✓ Saved: results/fast_training_history.png")

# Confusion Matrix
plt.figure(figsize=(10, 8))
cm = confusion_matrix(val_labels, val_preds)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=full_dataset.classes,
            yticklabels=full_dataset.classes,
            cbar_kws={'label': 'Count'},
            annot_kws={'fontsize': 16, 'fontweight': 'bold'})

plt.title(f'Confusion Matrix - Accuracy: {val_acc:.2f}%\n(Trained in {len(history["train_loss"])} epochs)', 
          fontsize=16, fontweight='bold', pad=20)
plt.ylabel('True Label', fontsize=13, fontweight='bold')
plt.xlabel('Predicted Label', fontsize=13, fontweight='bold')

# Add percentages
for i in range(len(cm)):
    for j in range(len(cm)):
        percentage = cm[i, j] / cm[i].sum() * 100
        plt.text(j+0.5, i+0.7, f'({percentage:.1f}%)',
                ha='center', va='center', fontsize=11, color='darkred', fontweight='bold')

plt.tight_layout()
plt.savefig('results/fast_confusion_matrix.png', dpi=300, bbox_inches='tight')
print("✓ Saved: results/fast_confusion_matrix.png")

# Save summary
summary = f"""
{'='*60}
FAST TRAINING RESULTS - 25 EPOCH TARGET
{'='*60}

Training Configuration:
- Total epochs: {len(history['train_loss'])}
- Batch size: {batch_size}
- Phase 1 (frozen): {phase_1_epochs} epochs at LR 3e-3
- Phase 2 (unfrozen): {len(history['train_loss'])-phase_1_epochs} epochs at LR 1e-4/5e-4

Results:
- Best validation accuracy: {best_val_acc:.2f}%
- Final validation accuracy: {val_acc:.2f}%
- Target (95%): {'✓ ACHIEVED!' if val_acc >= 95 else 'Not reached'}

Dataset:
- Training samples: {len(train_dataset)}
- Validation samples: {len(val_dataset)}
- Classes: {full_dataset.classes}

Per-Class Performance:
"""

per_class_acc = cm.diagonal() / cm.sum(axis=1) * 100
per_class_precision = [cm[i,i] / cm[:,i].sum() * 100 if cm[:,i].sum() > 0 else 0 
                       for i in range(len(cm))]
per_class_recall = per_class_acc

for i, class_name in enumerate(full_dataset.classes):
    summary += f"\n{class_name}:\n"
    summary += f"  Accuracy:  {per_class_acc[i]:.2f}%\n"
    summary += f"  Precision: {per_class_precision[i]:.2f}%\n"
    summary += f"  Recall:    {per_class_recall[i]:.2f}%\n"

summary += f"\n{'='*60}\n"

if best_val_acc >= 95:
    summary += "🎉 SUCCESS! Target achieved in under 25 epochs!\n"
else:
    summary += f"\nTo improve to 95%+:\n"
    summary += f"1. Train 5-10 more epochs (total 30-35)\n"
    summary += f"2. Enable GPU if not already (10-50x faster)\n"
    summary += f"3. Reduce augmentation if accuracy < 90%\n"
    summary += f"4. Use ensemble of 3 models (+2-5% boost)\n"

summary += f"{'='*60}\n"

with open('results/fast_training_summary.txt', 'w') as f:
    f.write(summary)
print("✓ Saved: results/fast_training_summary.txt")

print("\n" + summary)

print("\n" + "=" * 60)
print("TRAINING COMPLETE - ALL RESULTS SAVED")
print("=" * 60)
print("\nGenerated files:")
print("  📊 results/fast_training_history.png")
print("  🔢 results/fast_confusion_matrix.png")
print("  📄 results/fast_training_summary.txt")
print("  💾 best_fast_model.pth")
print("\n" + "=" * 60)
```
